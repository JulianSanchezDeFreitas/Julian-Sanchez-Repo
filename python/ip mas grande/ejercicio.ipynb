```python
import pandas as pd
import csv
from pathlib import Path

ruta =  Path("archivos") / "logs_catedra.csv"



def retorno_datos_usuarios(ruta:Path)->pd.DataFrame:
    data = pd.read_csv(
        ruta, sep=",", quoting=csv.QUOTE_MINIMAL, on_bad_lines="skip"
    )
    data = data[(data["Contexto del evento"] == "Página: Material de las clases") & (data["Nombre evento"] == "Módulo de curso visto") ].copy()
    data = data.sort_values(by="Dirección IP", ascending=False)

    return data
```

```python
datos_usuarios_df = retorno_datos_usuarios(ruta)

def ip_mas_grande(data:pd.DataFrame , cantidad:int = 5)->pd.Series:
    data = data.drop_duplicates(subset="Dirección IP").head(cantidad)
    return data[["Nombre completo del usuario","Dirección IP"]]
usuarios_ip_mas_grande_series = ip_mas_grande(datos_usuarios_df)

print(usuarios_ip_mas_grande_series)
```
